# DATA Simplification

1. Change the data set to action 1,3,7 (approval/denial only)

```python
import pandas as pd
import os
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
path=Path()
```

```python
parent_path=str(path.parent.absolute())[:-10]
```

```python
record=parent_path+"/data/raw/hmda_2017_ca_all-records_labels.csv"
df=pd.read_csv(record, low_memory=False)
```

```python
simpDF_road=parent_path+'/data/csvs/hmda_2017_ca_decisions_only.csv'
```

```python
#save_DF_with_only_1,3,7,2,8
simpDF=df[df['action_taken'].isin([1,3,7,2,8])]
simpDF.head()
```

```python
len(simpDF)
```

```python
simpDF.to_csv(simpDF_road, index=False)
```

2. Turn data set into Code and Numeric only

```python
removedDF=simpDF.drop(columns=[x for x in df.columns if "name" in x])
removedDF.head()
```

```python
removedDF.to_csv(parent_path+"/data/csvs/hmda_2017_ca_noname.csv", index=False)
```
